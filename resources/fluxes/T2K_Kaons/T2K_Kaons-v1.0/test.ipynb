```python
with open('TOT_MINUS_NUMUBAR.dat','r') as minus, open('TOT_PLUS_NUMU.dat','r') as plus, open('ratio.dat','w') as w:
    for lineMINUS, linePLUS in zip(minus.readlines(), plus.readlines()):
        E, minus = lineMINUS.strip().split()
        _, plus = linePLUS.strip().split()
        
        E = float(E)
        ratio = float(minus) / float(plus)
        
        print(E, ratio, file=w)
        
import matplotlib.pyplot as plt
```

```python
Es, a, b = [], [], []
with open('kaon-flux-numubar_bar.dat','r') as r:
    for line in r.readlines():
        E, ratio = line.strip().split()
        Es.append(float(E))
        a.append(float(ratio))
with open('kaon-flux-numu.dat','r') as r:
    for line in r.readlines():
        _, ratio = line.strip().split()
        b.append(float(ratio))
        
plt.plot(Es, a)
plt.plot(Es, b)

plt.yscale('log')
plt.show()
```
